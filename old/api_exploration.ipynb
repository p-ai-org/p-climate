```python
import requests
```

```python
# API URL to retrieve all of the NOAA Paleo Datasets that
# come from Coral/Sclerosponges and use d18O as a proxy
api_url = "https://www.ncei.noaa.gov/access/paleo-search/study/search.json?dataPublisher=NOAA&dataTypeId=4&cvWhats=chemical%20composition%3Eisotope%3Eisotope%20ratio%3Edelta%2018O&headersOnly=true"
```

```python
response = requests.get(api_url, timeout=5)
```

```python
print(response.url)
print(response.status_code)
```

```python
print(response.headers)
```

```python
data_dict = response.json()
data_dict
```
